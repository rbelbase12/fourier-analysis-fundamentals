```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 1000)
f = x  # sawtooth
approx = np.zeros_like(x)

for n in range(1, 50):
    approx += (2 * ((-1)**(n+1)) / n) * np.sin(n*x)

plt.plot(x, f, label='Original Function')
plt.plot(x, approx, label='Fourier Approximation (n=50)')
plt.legend()
plt.show()
```
